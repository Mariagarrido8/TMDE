**APRENDIZAJE MÁQUINA**

1) Recibir un conjunto de entrenamiento *train*: observaciones (muestras) y etiquetas.

2) Escoger un clasificador.

3) Seleccionar los hiperparámetros (optimización del modelo).

4) Ajustar el modelo con el conjunto de entrenamiento.

5) Recibir el conjunto de test *test*: observaciones sin etiquetas.

6) Predecir con el modelo una etiqueta para cada observación del conjunto de test.

7) Recibir las etiquetas reales del conjunto de test y evaluar la precisión del modelo.

Un conjunto de observaciones es un vector de muestras "x", y esas muestras sabemos a que clase pertenecen (conocemos su etiqueta, lo que se conoce como aprendizaje supervisado).


**CONJUNTO DE ENTRENAMIENTO Y DE TEST**

70% de los datos es para entrenamiento y el 30% de los datos es para test aproximadamente. Esta partición se hace para poder utilizar la partición de test para evaluar las prestaciones finales de
un modelo. Siempre tenemos que tener más datos para entrenamiento para poder ajustar lo máximo posible la máquina.

Con los datos que nos dan (matriz de datos "x" y vector columna "y"), tenemos que diseñar una máquina que sea capaz de clasificar los datos en clases. Las dimensiones de la matriz "x" serán las características de las clases, y el vector columna "y" es la clase de la muestra que se está pasando, es decir, contiene las etiquetas.

Si trabajamos con una base de datos binaria, tendremos solo dos clases en las que las etiquetas serán 0 o 1, dependiendo de si pertenece a una clase o a la otra.

**ESTANDARIZACIÓN**

La estandarización se utiliza para normalizar los datos.
Para normalizar, se hace por columnas restando la media y dividiendo por la desviación típica.

Al normalizar tenemos en cada característica media 0 y varianza 1 en train (no en test, porque se hace la media respecto de train).


**SOBREAJUSTE**

Si se acierta siempre con el conjunto de entrenamiento, se sobreajusta la máquina. Esto se da cuando hay mucha diferencia entre el resultado del conjunto de entrenamiento y el resultado del conjunto de test, es decir, que la máquina se ajusta tanto a las muestras que tenemos que no clasifica bien después las muestras reales. 

Que se produzca sobreajuste no es favorable, ya que una vez esté diseñada la máquina no se pasan las muestras del conjunto de entrenamiento, sino que se pasan muestras reales (las muestras de test) que son las que realmente el clasificador tiene que clasificar bien.

**HIPERPARÁMETROS**

Parámetros de la máquina de aprendizaje que debemos validar para obtener los resultados óptimos ajustándolos a nuestro problema. Por tanto, hay que probar varios valores para saber en cada problema cuál es el que se ajusta mejor.



**CONJUNTO DE TEST SEPARADO O VALIDACIÓN CRUZADA**

Mide el coste real de un clasificador, es decir, la tasa de error media.

Se hace validación cruzada para cada combinación de hiperparámetros. Divide el número de muestras en k conjuntos (k-fold) y coge uno de ellos para validar en cada iteración y el resto para entrenar (por ejemplo: en un 5-fold, en cada iteración tenemos 4 subconjuntos para train y 1 para validar).



**K-NN**

- Tiene un único hiperparámetro "k" (número de vecinos). El valor de "k" determina el tamaño del vecindario y localidad del clasificador.


- Compara los vecinos, y la clase mayoritaria de esos vecinos es la que se clasifica. Para compararlos, se calcula la distancia euclídea.

- K-NN, por tanto, en memoria es muy costoso porque hay que calcular todas las distancias a todas las muestras.

- K-NN ponderado: cuando tenemos empate entre clases, en k-NN ponderado se elige la muestra más cercana de entre los vecinos, a diferencia de k-NN normal que se elige la clase al azar. Por tanto, los vecinos tienen distinto peso según lo cercano que estén.

- El entrenamiento del algoritmo k-NN es rápido puesto que,para cada una de las muestras, directamente nos dice a qué clase pertenece calculando, de entre los k vecinos más próximos, cuál de las clases es la más frecuente. Por tanto, no hay mucho cómputo implicado. Sin embargo, la evaluación tarda lo mismo que usando cualquier otro método. En consecuencia, la evaluación es más lenta que el entrenamiento.

- K-NN, además, es muy sensible a la estandarización de los datos, porque vamos a utilizar las distancias a lo largo del entrenamiento, y si unas variables toman valores en un rango muy diferente a otras, la medición de distancias sería incoherente





**ÁRBOL DE DECISIÓN**

- Clasificador binario.

- Estructura jerárquica diseñada a partir de una aplicación recursiva de clasificadores *stump*. 

- Se define a partir de un umbral dividiendo las observaciones en dos: aquellas observaciones para las que el umbral es cierto y aquellas para las que el umbral es falso.

- Construcción:

  1) Nodo raíz (primer nodo del que parto).

  2) División del nodo raíz en nodos ramas con el algoritmo *stump* en el que cada rama recibe un subconjunto del conjunto de entrenamiento (el nodo *stump* es el nodo en el que se clasifican las clases).

  3) Nodos hojas final del árbol.

- Se puede utilizar en problemas multidimensionales, en la que para cada dimensión delvolverá el mejor *stump*.

- Métricas para medir la calidad en un árbol de decisión:

  - Tasa de clasificación de error.

  - Coeficiente de Gini.

  - Entropía cruzada.

- Hiperparámetros del árbol de decisión: máximo número de nodos hoja, calidad (coeficiente de giny o entropía cruzada), máxima profundidad del árbol, número de características exploradas en el *stump*, etc.

- En los árboles, el entrenamiento es lento porque supone mucho cálculo: debe distinguirse cada vez que se alcanza un nodo cuál es la variable más eficaz para ayudar en la clasificación de la muestra. Eso supone utilizar todas las variables no usadas hasta el momento y comparar. En consecuencia, el cómputo es mucho. Sin embargo, la evaluación sobre el rendimiento es rápida, puesto que para clasificar una muestra (una vez se tiene el modelo) basta con hacer unas comparaciones de valores de algunas de sus variables con un umbral.


**RANDOM FOREST**

- Métodos de conjunto: En vez de utilizar una máquina que está muy bien entrenada, utilizar varias máquinas peor entrenadas (aprendices débiles) en donde se combinan sus salidas.

- Random Forest es un conjunto de árboles de decisión.

- En cada árbol no se utilizan todas las muestras de entrenamiento, sino un subconjunto de muestras donde le decimos el tamaño que queremos. El subconjunto de entrenamiento es aleatorio.

- Puede que haya muestras que se repitan.

- Se evalúa en cada *stump* unas características diferentes. Siempre se evalúa el mismo número de dimensiones pero se eligen las características de manera aleatoria en cada *stump*.

- La salida mayoritaria con respecto a cada árbol es la que se elige para tomar la decisión final. 

- En RF tanto en entrenamiento como en evaluación es lento, ya que RF usa varios árboles.

**SVM**

- Máquinas de Vector de Soporte.

- Las SVMs separan de forma lineal un problema. Para 2D son rectas, para 3D es un plano.

- En SVMs maximizamos el margen, que es lo mismo que minimizar el vector de pesos.

- Se resuelve con programación cuadrática introduciendo multiplicadores de Lagrange. 

- Vectores soporte: multiplicadores de Lagrange distintos de 0. Los que son 0 no se necesitan para la clasificación porque no serán vectores soporte.



**SVM en casos no lineales**

- Casos en los que no se pueda resolver de forma lineal.

- Aumento de dimensiones: el hecho de que las observaciones no sean linealmente separables en el espacio original no significa que no lo sean en un espacio de mayores dimensiones. Cuantas más dimensiones haya, más complejo va a ser el algoritmo. La complejidad es infinita y hay que controlarla, que para ello se utilizan los kernels.

- Truco del kernel: una vez seleccionado un kernel, tenemos garantía de que existe un mapeo subyacente en un cierto espacio de características donde se pueden aplicar algoritmos que sólo impliquen calcular un producto escalar entre las observaciones. Tipos de kernel:

  - Kernel lineal: ninguna transformación, aplicamos los datos de forma lineal.
  
  - Kernel de Funciones de Base Radial (RBF)

  - Kernel polinómico

  - Combinación de kernels también son otro kernel


**SVC (SVM DE MARGEN SUAVE)**

- Los kernels tienen un riesgo de sobreajuste muy elevado por ejemplo cuando hay outliers, por lo que en estas situaciones sería mejor dejar que el algoritmo tolere algunos errores entre las muestras de entrenamiento para conseguir una frontera de clasificación más suave.

- "C": es un hiperparámetro que controla el trade-off entre la maximización del margen y los errores de clasificación en el conjunto de entrenamiento.

  - Las muestras correctamente clasificadas tendrán valor 0.

  - Las muestras mal clasificadas serán mayores que 1.

  - Las muestras dentro del margen estarán en el intervalo entre 0 y 1.

  Nota: para valores de "C" altos se da más importancia a clasificar correctamente todas las observaciones del conjunto de entrenamiento, y para valores de "C" bajos se da más importancia a maximizar el margen.


## 2.2. Estimation of hyperparameters with grid search

The performance of most of the machine learning algorithms depends strongly on the selected values for the hyperparameters. In the case of Random Forest for instance you will consider the number of trees, $k$, and the maximum number of leaves per tree as hyperparameters. 

The training (with method `fit`) of the RF grows optimally all the trees in the forest by finding optimal values for the binary tests performed in each branch node of each tree. But before calling `fit`, one needs to instantiate the RF with suitable values for all the hyperparameters (number of trees and number of leaves per tree in this case).

A common way of finding values for these hyperparameters is a **grid search** with **cross validation**.

### 2.2.1. Grids of hyperparameters

This method consists in to form a **grid** with a number of dimensions equal to the number of hyperparameters that one needs to optimize. The size of each dimension of the grid is equal to the number of values in the range of the corresponding hyperparameter. Notice that this method explores **discrete** ranges for each hyperparameter.

In the example of the Random Forest  we will explore the following ranges:

- `n_estimators` $ \in [10, 20, 50, 100, 200, 500, 1000]$
- `max_leaf_nodes` $ \in [5, 10, 20, 50]$

Notice these ranges determine a $7\times 4$ grid: 28 different combinations of pairs (`n_estimators`, `max_leaf_nodes`)  in the grid. 

In models that depend on a larger number of hyperparameters one has to be careful with the granularity of the ranges as the combinatorial explosion of the size of the grid can be hard to manage.

### 2.2.2. Cross validation 

Cross validation is a commonly used procedure in machine learning to simulate the effect of training a model with a set of data and evaluate its generalization capabilities as the performance in a **separate dataset**. 

The cross validation process involves the following steps:

- Randomly partition the training dataset in $N$ disjoint subsets of similar sizes. Each of this subsets is called **fold** in machine learning jargon. Hence the term **N-fold cross validation**.

- Let us suppose we have chosen $N=3$ folds. This means the training data has been split in three subsets: $(X_1, Y_1)$, $(X_2, Y_2)$ y $(X_3, Y_3)$. 

- Create an instance of the model with the corresponding hyperparameters. The cross validation follows with the execution of the following loop

    For $n=1,2,\dots,N$ iterations:  
    1. Choose $(X_n,Y_n)$ as **validation set** for iteration $n$
    2. Prepare a **training set** for iteration $n$ joining the rest of the subsets (excluding the validation set)
    3. Fit the model instance with the training set of step 2
    4. Evaluate  the model instance (method `score`) with the validation set of step 1
    5. Keep the *score* achieved in the $n$ iteration

- Once the loop is finished, we have $N$ scores, each corresponding to the evaluation of the model fitted in each iteration with the corresponding validation set.
- Estimate the **real score** that an instance of the model fitted using all the data would yield in a separate dataset computing the **mean** and **standard deviation** of the $N$ validation scores.

Typical values for the number of folds include $N\in \{3, 5, 10\}$


### 2.2.3. Cross validation to explore the grid

The grid is explored by a loop that visits all its nodes that runs a cross validation. The process details are the following:

For each node of the grid:  
1. Create an instance of the model with the hyperparameters set to the values that define that node of the grid
2. Run a **cross validation** to estimate the test performance that the model would yield if it were fitted using the values for hyperparameter that correspond to that node. 
3. Store the cross validation score for that node of the grid.

Once all the nodes of the grid have been cross validated, the procedure outputs the combination of hyperparameters that achieved the best performance in cross validation. 

### 2.2.4 Grid search in sklearn

There is a module in sklearn that implements this algorithm for exploring a grid of hyperparameters with cross validation: [`GridSearchCV`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html#sklearn.model_selection.GridSearchCV)


## 2.3. Evaluation of the training and testing times with cross_validate

Besides the classification accuracy there are other relevant measures that can help you decide for a machine learning method or another. These metrics include the **time needed to train the method** and the **time needed to evaluate a test set** with an already trained method.

An immediate way of estimating these metrics is to run the method several times and compute some statistics. Cross validation is a clean way to obtain more or less robust estimations of these quantities. Sklearn provides an stand-alone (besides the one in GridSearch) implementation of cross validation in module [`cross_validate`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_validate.html#sklearn.model_selection.cross_validate). This module outputs a very complete information, including:
  - `test_score`: result of the test of each validation *fold*, 
  - `fit_time`: time to fit each *fold*,
  - `score_time`: time to run the test in each fold.

## 3.1. Ranges for hyperparameters
- `n_neighbors` $\in \{1, 3, 5, 7, 10, 15, 20, 50, 100, 200, 500\}$

```python

#gridsearch para comparar todas las combinaciones posibles de los hiperparámetros y escoge la mejor de ellas
#gridsearchCV es lo mismo pero con validación cruzada
from sklearn.model_selection import GridSearchCV 

from sklearn.neighbors import KNeighborsClassifier

#hiperparámetros de knn: n_neighbors (número de vecinos) y weights (pesos)
knn_params = {'n_neighbors':[1,3,5,7,10,15,20,50,100,200,500], 'weights': ('distance','uniform')} #peso 'distance' significa knn ponderado

clf_knn = GridSearchCV(KNeighborsClassifier(), knn_params) #pasamos a gridsearchcv el clasificador y los hiperparámetros de éste

dat_no_s=clf_knn.fit(X_train, y_train) #ajustamos para los datos sin normalizar
knn_no_s_best_score = dat_no_s.best_score_
knn_no_s_best_params = dat_no_s.best_params_

dat_s = clf_knn.fit(X_train_s, y_train) #ajustamos para los datos normalizados
knn_s_best_params=dat_s.best_params_
knn_s_best_score=dat_s.best_score_

print("Best Parameters kNN without scaling")
print(knn_no_s_best_params)
print('Best Score kNN {0:.2f}%'.format(knn_no_s_best_score))
print("")
print("Best Parameters kNN with scaling")
print(knn_s_best_params)
print('Best Score kNN {0:.2f}%'.format(knn_s_best_score))

print("el mejor clasificador es el normalizado");

#Me quedo con el mejor estimador:
grid_knn=dat_s
```

**In this section you must use the training data partition exclusively. Consider the test data partition is not available yet.**

## 8.2 Alternative measures of performance: ROC and AUC

The spam email detection is a probem in which the misclassifications bear different consequences depending on the true class of the observation:

- A true spam email classified as non-spam will show in your inbox and it will take a few seconds of your time to fix it.
- A true non-spam email classified as spam will not show in your inbox, it will either be deleted or moved to a spam-email folder. It will take you from hours to several days to fix it.

- The classification error treats both sources of error equally. Therefore we can resource to alternative measures of performance.
